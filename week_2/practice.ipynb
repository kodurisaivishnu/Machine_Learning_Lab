```python
import pandas as pd

# Identifying price columns
price_cols = [col for col in df.columns if 'price' in col.lower()]
print(price_cols)  # Debugging step to check the identified columns

# Convert price columns to numeric
for col in price_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Sort DataFrame by the first price column
sorted_df = df.sort_values(by=[price_cols[0]])

# Get details of the least and most expensive cars
least_expensive = sorted_df.iloc[0:1, :]
most_expensive = sorted_df.iloc[-1:, :]

# Concatenating both rows properly
result = pd.concat([least_expensive, most_expensive])
display(result)

```

```python
#12. Write a function to find min and max values of any column 
# col = df.isna().sum().idxmax()
# df[col].fillna(df[col].mean())

def MinMax(col):
  return [df[col].min(),df[col].max()]
```

```python
# 13. Call the above function to find min max of horse  power, length, fuel efficiency 
cols = df.select_dtypes(include =[np.number]).columns
for col in cols:
  print(MinMax(col))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is already loaded with the car sales data
cols = ['Price in thousands', 'Sales in thousands', 'Horsepower', 'Fuel efficiency']

# Convert columns to numeric (handling errors)
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Set plot style
sns.set_style("whitegrid")

# Create subplots for multiple histograms
fig,axes = plt.subplots(2, 2)
# Plot histograms
for i, col in enumerate(cols):
    ax = axes[i // 2, i % 2]
    sns.histplot(df[col], bins=20, kde=True, ax=ax)
    ax.set_title(f"Distribution of {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Frequency")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# 18. Print correlation coefficient value of price and sales
import pandas as pd
df['Price in thousands'] = pd.to_numeric(df['Price in thousands'],errors='coerce')
df['Sales in thousands'] = pd.to_numeric(df['Sales in thousands'], errors='coerce')

# Compute correlation coefficient
correlation = df['Price in thousands'].corr(df['Sales in thousands'])
correlation
```

```python
# 19. Plot correlation of price and sales using scatterplot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'Price in thousands' and 'Sales in thousands' are numeric
df['Price in thousands'] = pd.to_numeric(df['Price in thousands'], errors='coerce')
df['Sales in thousands'] = pd.to_numeric(df['Sales in thousands'], errors='coerce')

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Price in thousands'], y=df['Sales in thousands'], color='blue', alpha=0.6)

# Add regression line
sns.regplot(x=df['Price in thousands'], y=df['Sales in thousands'], scatter=False, color='red')


plt.title("Scatter Plot of Price vs Sales", fontsize=14)
```
